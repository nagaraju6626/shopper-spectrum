```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("../data/online_retail.csv", encoding="ISO-8859-1")
```

```python
df = df.dropna(subset=['CustomerID'])
```

```python
df = df.dropna(subset=['Description'])
```

```python
df.isnull().sum()
```

```python

duplicate_count = df.duplicated().sum()

print("Number of duplicate rows:", duplicate_count)
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Count cancelled invoices
cancelled = df[df['InvoiceNo'].astype(str).str.startswith('C')]

print("Cancelled Invoices:", cancelled.shape[0])
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
print("Cancelled invoices remaining:",
      df['InvoiceNo'].astype(str).str.startswith('C').sum())
```

```python
print("Rows with Quantity <= 0:", (df['Quantity'] <= 0).sum())
```

```python
df = df[df['Quantity'] > 0]
```

```python
print("Rows with Quantity <= 0:",
      (df['Quantity'] <= 0).sum())
```

```python
print("Rows with UnitPrice <= 0:",
      (df['UnitPrice'] <= 0).sum())
```

```python
df = df[df['UnitPrice'] > 0]
```

```python
print("Rows with UnitPrice <= 0:",
      (df['UnitPrice'] <= 0).sum())
```

```python
print("Final Dataset Shape:", df.shape)

print("\nRemaining Invalid Records:")
print("Cancelled:", df['InvoiceNo'].astype(str).str.startswith('C').sum())
print("Quantity <= 0:", (df['Quantity'] <= 0).sum())
print("UnitPrice <= 0:", (df['UnitPrice'] <= 0).sum())
```

```python
print(df['InvoiceDate'].dtype)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(df['InvoiceDate'].dtype)
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df.to_csv("../data/cleaned_online_retail.csv", index=False)
```
